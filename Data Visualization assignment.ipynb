In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour  each data point with specific class

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
iris= datasets.load_iris()
```

```python
x= iris.data
y= iris.target
```

```python
fig= plt.figure(1,figsize=(4,3))
ax= Axes3D(fig=fig,rect=[0,0,0.89,0.95],elev=42)

#Decomposition
pca= decomposition.PCA(n_components=3)

X= pca.fit_transform(x)

for name,species in [('Setosa',0),('Versicolor',1),('Virginica',2)]:
    ax.text3D(X[y == species, 0].mean(),
              X[y == species, 1].mean()+1.5, 
              X[y == species, 2].mean(),name,horizontalalignment='center')


#Scattering the species data in 3D plot:
y = np.choose(y,[1,2,0])
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y)

plt.show()
```
