```python
import numpy as np
from sklearn.cluster import KMeans
import scipy.misc
import matplotlib.pyplot as plt
```

```python
#Visualize
f = scipy.misc.face(gray=True)
plt.figure(figsize=(10, 5))
plt.imshow(f, cmap=plt.cm.gray)
plt.show()
```

```python
# Compressing the gray scale image into 5 clusters
rows = f.shape[0]
cols = f.shape[1]
comp = f.reshape(rows*cols,1)
```

```python
#Dividing the image into 5 clusters
kmeans = KMeans(n_clusters = 5)
kmeans.fit(comp)
```

```python
o= kmeans.labels_
```

```python
label= o.reshape(rows,cols)
```
