Problem 2:

How many Unique patterns that exist in the historical stock data set, based on
fluctuations in price.

```python
#The dataset above is arranged in terms of price at different timestamp for different stocks
```

```python
#We have 246 to 256 trading days. The median no of days is 253. We can calculate the returns from the stock price wrt to previous rows

returns= stock1.pct_change().mean()*253
returns= pd.DataFrame(returns,columns=['Returns'])
```

```python
returns
```

```python
from math import sqrt
returns['Volatility']= stock1.pct_change().std()*sqrt(253)
```

```python
returns
```

```python
l1= np.asarray(returns.Returns)
```

```python
l2 =np.asarray(returns.Volatility)
```

```python
l3= np.asarray([l1,l2])
```

```python
performance= l3.T
```

```python
from sklearn.cluster import KMeans
```

```python
optimum =[]
for k in range(2,20):
    k_means = KMeans(n_clusters = k)
    k_means.fit(performance)
    optimum.append(k_means.inertia_)
```

```python
plt.plot(range(2,20),optimum)
plt.show()
```

from the above elbow curve, the optimum of no of clusters can be taken as 5

```python
k_means.cluster_centers_
```

```python
km =KMeans(n_clusters=5)
```

```python
km.fit(performance)
```

```python
distance = km.inertia_
```

```python
print(distance)
```

```python
km.cluster_centers_
```

```python
clusters= km.cluster_centers_
```

```python
label=km.labels_
label
```

```python
from scipy.cluster.vq import kmeans,vq
```

```python
from pylab import rcParams, plot,show
```

```python

centers,_= kmeans(performance, 5)
# Assign each sample to a cluster
idx,_ = vq(performance, centers)

# some plotting using numpy's logical indexing
plot(performance[idx==0,0],performance[idx==0,1],'ob',
     performance[idx==1,0],performance[idx==1,1],'oy',
     performance[idx==2,0],performance[idx==2,1],'or',
     performance[idx==3,0],performance[idx==3,1],'og',
    performance[idx==4,0],performance[idx==4,1])
plot(centers[:,0],centers[:,1],'sg',markersize=8)
show()
```

Problem 3:

Identify which all stocks are moving together and which all stocks are different from
each other.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
st= StandardScaler()
```

```python
stock_std= st.fit_transform(stock1.values)
```

```python
stock_std
```

```python
pca = PCA(n_components =50)
```

```python
comp= pca.fit(stock_std)
```

```python
variance = comp.explained_variance_ratio_
```

```python
variance1= np.cumsum(variance)
```
